```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_excel('Qnre.xlsx', header=0)
df2=df.copy()
```

```python
#This code block generates the responses for multiple respose questions as lists
sata_types = ['Select All That Apply', 'Open Ended-Select All That Apply'] # create a list of multiple response question types
for ind, value in  enumerate(df2['Type']):
    df2.at[ind, 'vars'] = re.findall('^\n?\d+\)\s*[^\n]*',
                                     df2['Questions'].at[ind],
                                     flags = re.MULTILINE) if value in sata_types else df2['Name'].at[ind]

    print(ind, df2.at[ind, 'vars'])

df2 = df2.explode('vars') # unpack the lists in the vars column so that each value is a row
df2['vars'] = [re.sub('\d\)\s*', '', x) for x in df2['vars']] #remove the numbering from the options

print('\n', df2['vars'])

#Rename the multiple response variables using the question name as the prefix
df2['vars'] = np.where((df2['vars'] != df2['Name'] ), df2['Name']+'_'+df2['vars'], df2['Name'])
```

```python
df2
```

```python
variables = df2['vars'].tolist() #convert the variables to a list 

#unwanted variables
variables_to_drop = ['Refusal', 'Ineligible', 'Close-out-Incentive']

variables = [var for var in variables if var not in variables_to_drop]

variables
```

```python
dataset.drop(drop_columns, axis=1, inplace = True)
dataset
```
